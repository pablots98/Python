```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

Step 1: Set up the 2D grid and the initial position of the agents 

```python
class Agent:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.active = True  # Agent can move unless it interacts

    def move(self):
        if self.active:  # Move only if the agent is active
            move_direction = np.random.choice(['up', 'down', 'left', 'right'])
            if move_direction == 'up' and self.y < 99:
                self.y += 1
            elif move_direction == 'down' and self.y > 0:
                self.y -= 1
            elif move_direction == 'left' and self.x > 0:
                self.x -= 1
            elif move_direction == 'right' and self.x < 99:
                self.x += 1
```

Step 2: Initialize agents randomly in the grid

```python
def initialize_agents(num_agents):
    agents = []
    for _ in range(num_agents):
        x, y = np.random.randint(0, 100, size=2)
        agents.append(Agent(x, y))
    return agents
```

Step 3: Calculate the distance between two agents

```python
def distance(agent1, agent2):
    return sqrt((agent1.x - agent2.x) ** 2 + (agent1.y - agent2.y) ** 2)
```

Step 4: Interaction rule: stop agents withing a distnace of 1

```python
def check_interactions(agents):
    for i in range(len(agents)):
        for j in range(i + 1, len(agents)):
            if distance(agents[i], agents[j]) <= 1:
                agents[i].active = False
                agents[j].active = False
```

Step 5: Simulation of the system over 100 time steps

```python
def simulate(agents, steps):
    for _ in range(steps):
        for agent in agents:
            agent.move()
        check_interactions(agents)
```

Step 6: Visulatization of agents 

```python
def plot_agents(agents, step):
    x_coords = [agent.x for agent in agents]
    y_coords = [agent.y for agent in agents]
    colors = ['red' if not agent.active else 'blue' for agent in agents]  # Red for stopped agents
    plt.scatter(x_coords, y_coords, c=colors)
    plt.xlim(0, 100)
    plt.ylim(0, 100)
    plt.title(f"Agent Positions at Step {step}")
    plt.show()
```

Step 7: Run the simulation and visualize every 10 steps 

```python
agents = initialize_agents(50)
steps = 100

for step in range(0, steps + 1, 10):
    simulate(agents, 10)
    plot_agents(agents, step)
```
